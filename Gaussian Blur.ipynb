```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2 
image = cv2.imread('Desktop/Udacity/brain.jpg')
#plt.imshow(image)

%matplotlib inline 
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
print(image_copy.shape)
```

```python
#convert image to grey scale 

grey= cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

#plt.imshow(grey)
grey_f = cv2.GaussianBlur(image_copy, (5,5), 0)

f, (ax1,ax2) = plt.subplots(1, 2 , figsize=(20,10))


ax1.set_title('Original')
ax1.imshow(image_copy)

ax2.set_title('Gaussian')
ax2.imshow(grey_f)

```

```python
sobel_x  = np.array([[ -1, 0 ,1 ],
                     [ -2, 0, 2],
                     [ -1, 0, 1 ]])
f_1 = cv2.filter2D(image_copy, -1, sobel_x)
f_2 = cv2.filter2D(grey_f, -1, sobel_x)

f, (ax1,ax2) = plt.subplots(1, 2 , figsize=(20,10))


ax1.set_title('Original')
ax1.imshow(f_2)

ax2.set_title('Gaussian')
ax2.imshow(f_1)
```
